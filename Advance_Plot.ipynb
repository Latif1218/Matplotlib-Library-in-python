##### a example with 2 array

```python
x = [1,2,3,4]
y = [5,6,7,8]
plt.scatter(x,y)

plt.text(1,5,'point 1')
plt.text(2,6,'point 2')
plt.text(3,7,'point 3')
plt.text(4,8,'point 4',fontdict = {'size':12,'color':'brown'})
```

#### so thais the way to spacify ...... so solve the previus problem

```python
plt.figure(figsize=(18,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'])

for i in range (sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

##### dot size increses with batsman run value as call the size( s = ) function

```python
plt.figure(figsize=(18,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])

for i in range (sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

#### plot horizontal and vertical line

```python
plt.figure(figsize=(18,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])

plt.axhline(130,color='red')
plt.axhline(140,color='green')
plt.axvline(30,color='red')

for i in range (sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

##### ->plt.subplots() have two object (1. figure , 2. axis) and using axis we have to plot a graph

```python
plt.subplots()

# this = (<Figure size 640x480 with 1 Axes>, <Axes: >)
```

```python
fig, axis = plt.subplots(figsize = (15,6))

axis.scatter(batters['avg'],batters['strike_rate'])
axis.set_title('avg vs strike_rate')
axis.set_xlabel('avg')
axis.set_ylabel('strike_rate')
```

### so haw to plot side by side multiple acatter plot

```python
fig , ax =  plt.subplots(nrows=2,ncols=1,sharex = True,figsize = (10,6))

ax[0].scatter(batters['avg'],batters['strike_rate'],color = 'red')
ax[1].scatter(batters['avg'],batters['runs'])

ax[0].set_title('avg vs strike_rate')
ax[0].set_ylabel('strike_rate')

ax[1].set_title('avg vs runs')
ax[1].set_ylabel('runs')
ax[1].set_xlabel('avg')
```

```python
fig , ax =  plt.subplots(nrows=2,ncols=2,figsize = (10,10))

ax[0,0].scatter(batters['avg'],batters['strike_rate'],color = 'red')
ax[0,1].scatter(batters['avg'],batters['runs'])
ax[1,0].hist(batters['avg'])
ax[1,1].hist(batters['runs'])

ax[0,0].set_title('avg vs strike_rate')
ax[0,0].set_ylabel('strike_rate')

ax[0,1].set_title('avg vs runs')
ax[0,1].set_ylabel('runs')
ax[0,1].set_xlabel('avg')
```

### the third way to plot the subplote

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)  #(2 rows 2 cols 1st graoh)
ax1.scatter(batters['avg'],batters['strike_rate'])

ax2 = fig.add_subplot(2,2,2)  #(2 rows 2 cols 2nd graph)
ax2.scatter(batters['avg'],batters['runs'])

ax3 = fig.add_subplot(2,2,3)  #(2 rows 2 cols 3rd graph)
ax3.hist(batters['avg'])

ax4 = fig.add_subplot(2,2,4)  #(2 rows 2 cols 4th graph)
ax4.hist(batters['runs'])
```

#### 3D surface plot

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx, yy = np.meshgrid(x,y)
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize =(12,8))

ax = plt.subplot(projection = '3d')
p = ax.plot_surface(xx,yy,z,cmap = 'viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize =(12,8))

ax = plt.subplot(projection = '3d')
p = ax.plot_surface(xx,yy,z,cmap = 'viridis')
fig.colorbar(p)
```

#### Contour Plot

```python
fig = plt.figure(figsize =(12,8))

ax = plt.subplot()
p = ax.contour(xx,yy,z,cmap = 'viridis')
fig.colorbar(p)
```

#### contour f 2-D

```python
fig = plt.figure(figsize =(12,8))

ax = plt.subplot()
p = ax.contourf(xx,yy,z,cmap = 'viridis')
fig.colorbar(p)
```
