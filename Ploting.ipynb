##### from a pandas dataframe

```python
batsman = pd.read_csv('sharma-kohli.csv')
batsman
plt.plot(batsman['index'],batsman['V Kohli'])
```

##### plotting multiple plots

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

##### label title

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

plt.title('Rohit sharma vs virat kohli career comparison')
plt.xlabel('Season of IPL')
plt.ylabel('Runs Scored')
```

##### Colores(hex) and line(width and style) and marker(size)

```python
plt.plot(batsman['index'],batsman['V Kohli'],color = 'green',linestyle = 'dashed')
plt.plot(batsman['index'],batsman['RG Sharma'],color = 'black')

plt.title('Rohit sharma vs virat kohli career comparison')
plt.xlabel('Season of IPL')
plt.ylabel('Runs Scored')
```

##### linestyle and linewidth and marker ando also markersize

```python
plt.plot(batsman['index'],batsman['V Kohli'],color = '#c1ff00',linestyle = 'dotted',linewidth = '3',marker = 'D')
plt.plot(batsman['index'],batsman['RG Sharma'],color = '#ff0070',linestyle = 'dashdot',linewidth = '4',marker = 'o',markersize = 7)

plt.title('Rohit sharma vs virat kohli career comparison')
plt.xlabel('Season of IPL')
plt.ylabel('Runs Scored')
```

##### legend -> location

```python
plt.plot(batsman['index'],batsman['V Kohli'],color = '#c1ff00',linestyle = 'dotted',linewidth = '3',marker = 'D',label ='V Kohli')
plt.plot(batsman['index'],batsman['RG Sharma'],color = '#ff0070',linestyle = 'dashdot',linewidth = '4',marker = 'o',markersize = 7,label = 'RG Sharma')

plt.title('Rohit sharma vs virat kohli career comparison')
plt.xlabel('Season of IPL')
plt.ylabel('Runs Scored')
plt.legend()
```

##### Multiple Bar chart

```python
df = pd.read_csv('batsman_season_record.csv')
```

```python
plt.bar(np.arange(df.shape[0])-.2,df['2015'],width = .2,color = 'yellow',label  = '2015')
plt.bar(np.arange(df.shape[0]),df['2016'],width = .2,color = 'red',label = '2016')
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width = .2,color = 'blue',label = '2017')

plt.xticks(np.arange(df.shape[0]),df['batsman'])

plt.title('batsman vs season run plot')
plt.xlabel('Batsman name')
plt.ylabel('Season run')

plt.legend()
plt.show()
```

```python
##### a problem
###### 1. if x axis catagorical name is more large then we hav to solve this problem like this 
```

```python
children = [10,20,40,15,30]
colors = ['red red red red red red','blue blue blue blue blue blue','green green green green green green','yellow yellow yellow yellow yellow yellow','pink pink pink pink pink pink']           

plt.bar(colors,children,color ='green')
plt.xticks(rotation = 'vertical')
plt.show()
```

##### Stacked Bar chart

```python
plt.bar(df['batsman'],df['2015'],label = '2015')
plt.bar(df['batsman'],df['2016'],bottom = df['2015'],label = '2016')
plt.bar(df['batsman'],df['2017'],bottom = df['2016']+df['2015'],label = '2017')

plt.legend()
plt.show()
```

### Pie Chart

#### > Univariate/Bivariate Analysis
#### > Categorical vs numerical
#### > Use case -> To fins contibution on a standard scale

##### simple data pie chart

```python
data = [23,45,100,20,49]
subject = ['eng','science','maths','sst','Bangla']

plt.pie(data,labels=subject)
plt.show()
```

##### pie chart using proper dataset

```python
df = pd.read_csv('gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'],labels = df['batsman'],autopct = '%0.1f%%')
```

##### Explode shadow

```python
plt.pie(df['batsman_runs'],labels = df['batsman'],autopct = '%0.1f%%',explode=[0.1,0,0,0,0,0.3],shadow = True)
plt.show()
```
